```python
importlib.reload(a_v)

from multiprocessing import Process, Pool
import multiprocessing.managers
import tqdm

import warnings
warnings.filterwarnings("ignore")

in_Path = a_v.input_dir_extratracks

channels = ['...', '...', '...', '...', '...', '...','...', '...', '...','...','...']
channels_2 = ['.........', 
              '.........', 
              '.........', 
              '.........', 
              '.........', 
              '.........',
              '.........', 
              '.........', 
              '.........',
              '.........',
              '.........']

channels_= {}
j=0
for i in channels:
    channels_[i] = channels_2[j]
    j+=1

    

nomeTree = '...'
sample_extra = '...'
sample = '...'

def merger(channel):
    root_file = f"{sample}_{channels_[channel]}.parq"
    root_file_extra = f"{sample_extra}_{channel}.root"
    
    print('Reading the root file', root_file)
    df1 = pd.read_parquet(in_Path + root_file, engine='pyarrow')
    df1 = df1.drop('...', axis=1)

    print('Reading the second root file')
    if channel == 'offres' or channel == 'data':
        df2 = rp.read_root(in_Path + root_file_extra, 
                           key = '...',
                           columns = a_v.extraG_var_data)
    else:
        df2 = rp.read_root(in_Path + root_file_extra, 
                           key = '...',
                           columns = a_v.extraG_var)

    df4 = pd.DataFrame()

    print('Merging')
    if channel == '...' or channel == '...':
        df4 = df1.join(df2.set_index(['__experiment__','__run__','__event__']), on=['__experiment__','__run__','__event__'], how='left')
    else:
        df4 = df1.join(df2.set_index(['__event__', '__production__']), on=['__event__', '__production__'], how='left')
    
    print(root_file, '->before dropping duplicates', len(df4))
    df4 = df4.drop_duplicates()
    print(root_file, '->after dropping duplicates', len(df4))
    df4['...'] = df4['...'].replace(np.nan, 0)
    df4['...'] = df4['...'].replace(np.nan, 0)

    print('Saving the new files')

    df4.to_root(in_Path + root_file.replace('tight','extraG_tight').replace('.parq','.root'), key = nomeTree)
    df4.to_parquet(in_Path + root_file.replace('tight','extraG_tight'), compression='GZIP')
    print(root_file, 'done')

n_cpus = 2

pool = Pool(n_cpus)
    
run_list = channels
pool_result = pool.map(merger, run_list)

pool.close()
pool.join()
```
